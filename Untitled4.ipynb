```python
!pip install SpeechRecognition
!pip install pandas
# Import necessary libraries
import os
import speech_recognition as sr
import pandas as pd
from google.colab import drive

# Mount Google Drive to Colab
drive.mount('/content/drive')

# Set up paths and filenames
audio_path = '/content/drive/MyDrive/FYP/Datasets/Audio Datasets'
text_path = '/content/drive/MyDrive/FYP/Datasets/TextDatasets.txt'
csv_path = '/content/drive/MyDrive/FYP/Datasets/CsvDatasets.csv'

# Create empty dataframe to store text data
df = pd.DataFrame(columns=['text', 'label'])
#sfsd 
from google.colab import drive
drive.mount('/content/drive')
# Loop through labeled folders in the audio directory
```

```python

for label_folder in os.listdir(audio_path):
    if os.path.isdir(os.path.join(audio_path, label_folder)):
        label = label_folder
        label_path = os.path.join(audio_path, label_folder)
        
        # Loop through audio files in the label folder and convert them to text files
        for audio_file in os.listdir(label_path):
            if audio_file.endswith('.wav'):
                text_file = os.path.join(text_path, label, audio_file[:-4] + '.txt')
                audio_file = os.path.join(label_path, audio_file)
                r = sr.Recognizer()
                with sr.AudioFile(audio_file) as source:
                    audio_data = r.listen(source)
                    text = r.recognize_google(audio_data, language = 'en-IN', show_all = True)
                    df = df.append({'text': text, 'label': label}, ignore_index=True)
                    '''with open(text_file, 'w') as f:
                        f.write(text)'''
                
                # Append text and label to dataframe
                '''with open(text_file, 'r') as f:
                    text = f.read()'''
                

# Save dataframe to CSV file
df.to_csv(csv_path, index=False)
```
